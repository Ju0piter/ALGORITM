```python
import numpy
m = 5
n = 5
min_limit = -250
max_limit = 1000+3
randomnaya_matritsa = numpy.random.random_integers(min_limit,max_limit, size=(n,m))
print(randomnaya_matritsa)
```

```python
def per(matrixx):
    n=len(matrixx)
    m=len(matrixx[0])
    matrline=[0]*(n*m)
    x=int(0)
    for i in range(0, n):
        for j in range(0, m):
            matrline[x]=matrixx[i][j]
            x=x+1
    return matrline

def vivod(matrixx, matrline):
    n=len(matrixx)
    m=len(matrixx[0])
    x=int(0)
    matrix=[[0] * m for i in range(n)]
    for i in range(0, n):
        for j in range(0, m):
            matrix[i][j]=matrline[x]
            x=x+1
    return matrix 

#Cортировка выбором
def sortviborom(matrixx):
    matrline=per(matrixx)
    for i in range(0, len(matrline)):
        for j in range(i, len(matrline)):
            if matrline[i]>matrline[j]:
                matrline[i], matrline[j]=matrline[j], matrline[i]     
    return vivod(matrixx, matrline)
#Конец сортировки выбором

#Сортировка вставками
def sortvstavkami(matrixx):
    matrline=per(matrixx)
    for i in range(1, len(matrline)):
        j=i-1
        while (j>=0) and (matrline[j]>matrline[j+1]):
            matrline[j], matrline[j+1]=matrline[j+1], matrline[j]
            j=j-1
    return vivod(matrixx, matrline)     
#Конец сортировки вставками

# Сортировка Обменом
def sortobmen(matrixx):
    matrline=per(matrixx)
    l=len(matrline)
    while l>1:
        for i in range(l-1):
            if matrline[i]>matrline[i+1]:
                matrline[i], matrline[i+1]=matrline[i+1], matrline[i]
            l=l-1
    return vivod(matrixx, matrline)
# Конец сортировки обменом

#Сортировка Шелла
def shellsort(matrixx):
    matrline=per(matrixx)
    length=int(n*m)-1
    l=length
    while l>0:
        for i in range(l, length+1, 1):
            j=i
            while ((j-l)>=0) and (matrline[j-l]>matrline[j]):
                matrline[j], matrline[j-l]=matrline[j-l], matrline[j]
                j-=l
        l=int(l/2)
    return vivod(matrixx, matrline) 
#Конец сортировки Шелла

#Турнирная сортировка
def tyrsort(matrixx):
    matrline=per(matrixx)
    def _tyrsort(matrline):
        tree = [None] * 2 * (len(matrline) + len(matrline) % 2)
        index = len(tree) - len(matrline) - len(matrline) % 2
        for i, v in enumerate(matrline):
            tree[index + i] = (i, v)
        for j in range(len(matrline)):
            n = len(matrline)
            index = len(tree) - len(matrline) - len(matrline) % 2
            while index > -1:
                n = (n + 1) // 2
                for i in range(n):
                    i = max(index + i * 2, 1)
                    if tree[i] != None and tree[i + 1] != None:
                        if tree[i][1] < tree[i + 1][1]:
                            tree[i // 2] = tree[i]
                        else:
                            tree[i // 2] = tree[i + 1]
                    else:
                        tree[i // 2] = tree[i] if tree[i] != None else tree[i + 1]
                index -= n

            index, x = tree[0]
            matrline[j] = x
            tree[len(tree) - len(matrline) - len(matrline) % 2 + index] = None
    _tyrsort(matrline)
    return vivod(matrixx, matrline) 

#Конец турнирной сортировки

#Быстрая сортировка
def qs(matrixx):
    matrline=per(matrixx)
    def quicksort(matrline, first, last):
        i=first
        j=last
        x=matrline[int((first+last)/2)]
        while i<=j:
            while matrline[i]<x:
                i+=1
            while matrline[j]>x:
                j-=1
            if i<=j:
                if matrline[i]>matrline[j]:
                    matrline[i], matrline[j]=matrline[j], matrline[i]
                i+=1
                j-=1
        if i<last:
            quicksort(matrline, i, last)
        if first<j:
            quicksort(matrline, first, j)
    quicksort(matrline, 0, (n*m)-1)
    return vivod(matrixx, matrline)
#Конец быстрой сортировки

#Пирамидальная сортировка
def heapSort(matrixx):
    matrline=per(matrixx)
    def downHeap(matrline, k, n):
        el = matrline[k]
        while 2*k+1 < n:
            child = 2*k+1
            if child+1 < n and matrline[child] < matrline[child+1]:
                child += 1
            if el >= matrline[child]:
                break
            matrline[k] = matrline[child]
            k = child
        matrline[k] = el
  
    size = len(matrline)
    for i in range(size//2-1,-1,-1):
        downHeap(matrline, i, size)
    for i in range(size-1,0,-1):
        temp = matrline[i]
        matrline[i] = matrline[0]
        matrline[0] = temp
        downHeap(matrline, 0, i)
    return vivod(matrixx, matrline)
#Конец пирамидальной сортировки

import numpy
n=int(input())
m=int(input())

matrixx = numpy.random.random_integers(0, 10, size=(n, m))
print(*matrixx)
matrix=sortviborom(matrixx)
matrix1=sortvstavkami(matrixx)
matrix2=sortobmen(matrixx)
matrix3=shellsort(matrixx)
matrix4=tyrsort(matrixx)

#Быстрая сортировка
matrix5=qs(matrixx)
#
matrix6=heapSort(matrixx)
print("\nВывод сортировки выбором\n", *matrix)
print("\nВывод сортировки вставкой\n", *matrix1)
print("\nВывод сортировки обменом\n", *matrix2)
print("\nВывод сортировки Шелла\n", *matrix3)
print("\nВывод турнирной сортировки\n", *matrix4)
print("\nВывод быстрой сортировки\n", *matrix5)
print("\nВывод пирамидальной сортировки\n", *matrix6)
```
